## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
#pd.merge merging on "Mouse ID" column
mouseIDmerge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
mouseIDmerge_df.head(10)
```

```python
# Checking the number of mice.
mouseIDmerge_df["Mouse ID"].nunique()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleanmouseIDmerge_df = mouseIDmerge_df.loc[(mouseIDmerge_df['Mouse ID'] != 'g989')]
cleanmouseIDmerge_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
cleanmouseIDmerge_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#'Drug Regimen'
drugandtumor2_df = cleanmouseIDmerge_df[['Drug Regimen', 'Tumor Volume (mm3)']]
drugandtumor_df.head()

drugandtumor_df=drugandtumor2_df.groupby(['Drug Regimen'])

drugandtumor_df.head()

mean = drugandtumor_df['Tumor Volume (mm3)'].mean()
median = drugandtumor_df['Tumor Volume (mm3)'].median()
var = drugandtumor_df['Tumor Volume (mm3)'].var()
std = drugandtumor_df['Tumor Volume (mm3)'].std()
sem = drugandtumor_df['Tumor Volume (mm3)'].sem()


summary_df = pd.DataFrame({'mean': mean, 'median': median, 'var': var, 'std': std, 'sem': sem})
summary_df

summary_df['mean'] = summary_df['mean'].map('{:.2f}'.format)
summary_df['median'] = summary_df['median'].map('{:.2f}'.format)
summary_df['var'] = summary_df['var'].map('{:.2f}'.format)
summary_df['std'] = summary_df['std'].map('{:.2f}'.format)
summary_df['sem'] = summary_df['sem'].map('{:.2f}'.format)
summary_df


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# summary_df
# Using the aggregation method, produce the same summary statistics in a single line
onelinedsummary_df = cleanmouseIDmerge_df.agg({"Tumor Volume (mm3)": ["mean","median","var","std","sem"]})
onelinedsummary_df
```
